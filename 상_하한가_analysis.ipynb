```
upperdown = pd.read_excel('drive/My Drive/stockanalysis/upperdown.xlsx',header=0)
upperdown
```

```
upper_num = upperdown['상한가 종목들 중복 제거'].count()
under_num = upperdown['하한가 종목들 중복 제거'].count()
```

```
upper_names = upperdown['상한가 종목들 중복 제거'].iloc[0:upper_num]
under_names = upperdown['하한가 종목들 중복 제거'].iloc[0:under_num]
#upper_data = pd.merge(upper_name, "0"*upper_num.T)
upper_names
```

```
ucolumns = ['상위 10위 종목','상위 10위 등극 횟수']
ncolumns = ['하위 10위 종목', '하위 10위 등극 횟수']
#upper_sum = upper_names

upper_sum = pd.DataFrame()

uanalysis = pd.DataFrame(columns=ucolumns)
uanalysis['상위 10위 종목'] = upper_names

nanalysis = pd.DataFrame(columns=ncolumns)
nanalysis['하위 10위 종목'] = under_names

d = 0
for i in upper_names : 
  uanalysis.iloc[d,1] = sum(upperdown['상한가'] == i)
  d = d+1

t = 0
for i in under_names : 
  nanalysis.iloc[t,1] = sum(upperdown['하한가'] == i)
  t = t+1
```

```
sort_nan = nanalysis.sort_values(by='하위 10위 등극 횟수',ascending=False)
sort_uan = uanalysis.sort_values(by='상위 10위 등극 횟수',ascending=False)
```
